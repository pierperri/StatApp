```python
import pandas as pd
import numpy as np
import chardet
import math
```

```python
file = "database-imputed ESG.csv"
```

```python
#encoding du fichier csv
chardet.detect(open(file,'rb').read(100000))
```

```python
#Importation
data = pd.read_csv(file, encoding='Windows-1252')
```

```python
data
```

```python
col = data.columns
col
```

```python
#Est-ce qu'il y a un NaN?
for i in col:
    print(i,"        ",any(np.array(data[i].isna())))
#Oui
```

```python
#Où sont les NaN?
a,b = data['isin'],data['business_desc']
for i in range(len(a)):
    try:
        if math.isnan(a[i]):
            print('a',i, a[i])
    except:
        pass
    try:
        if math.isnan(b[i]):
            print('b',i, b[i])
    except:
        pass
```
